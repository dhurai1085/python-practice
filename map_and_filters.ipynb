```python
countries_list = ['india','german','japan','russia']
for x in countries_list:
  print(len(x))
```

```python
countries_len = list(map(len,countries_list))
countries_len
```

```python
def return_len(x):
  return f'{x} : {len(x)}'
```

```python
list(map(return_len,countries_list))
```

```python
mixed_list = ['argentina','malaysia','indonesia',88]
```

```python
def is_str(x):
  return type(x)==str
```

```python
print(f' Map:{list(map(is_str,mixed_list))}')
print(f'filter:{list(filter(is_str,mixed_list))}')
```
