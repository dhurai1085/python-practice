```python
# Create a list and print it
num_list = [2,44,55,55,6,71,17,17,3]
print(f'List elements:\n{num_list}')

# Find the length of a list
length = len(num_list)
print(f'Length of the num_list:\n{length}')

# Find the sum of elements using built-in function
sumValue = sum(num_list)
print(f'Sum of all the values in num_list:\n{sumValue}')

# Find the maximum element &  the minimum element
minValue,maxValue = min(num_list),max(num_list)

print(f'Max value: {maxValue}- Min value: {minValue}')

# Add an element using append()
num_list.append(14)
print(f'after adding a values to list:\n num_list')

# Insert an element at a specific position using insert()
num_list.insert(2,0)
print(f'after inserting a element into list:\n{num_list}')

# Remove an element using remove()
num_list.remove(14)
print(f'After removing an element from list:\n{num_list}')

# Remove last element using pop()
num_list.pop()
print(f'After removing last element:\n{num_list}')

# Sort a list using sort()
num_list.sort()
print(f'Sorted list:\n{num_list}')

# Reverse a list using reverse()
num_list.reverse()
print(f'Reversed list:\n{num_list}')

# Find second largest using sorted()
sorted(num_list,reverse=True)
print(f'Second largest number in list:\n{num_list[1]}')

# Get first 3 elements using slicing
first_three = num_list[0:3]
print(f'first 3 elements in the list:\n{first_three}')

# Get last 3 elements using slicing
last_three = num_list[-3::]
print(f'last three elements in list:\n{last_three}')

# Reverse list using slicing
reversed_list = num_list[::-1]
print(f'Reversed list:\n{reversed_list}')

# Extract alternate elements
alternate_numbers = num_list[::2]
print(f'Alternate numbers on the list:\n{alternate_numbers}')
```
