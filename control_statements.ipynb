```python
# finding valid email id's

email_ids = ['w3school@gamil.com','adc@gamil.com','example.com','karthee@gmail.com']
print('valid email ids:')
for email in email_ids:
  if '@' not in email:
    continue
  username,domain = email.split('@',1)
  if len(username) < 4 or '.' not in domain:
       continue
  print(email)
```

```python
# list of squares for first n natural numbers

n = 20
sqrt_list = []
for n in range(1,n+1):
  sqrt_list.append(n**2)
print(sqrt_list)

print('after using list comprehension:')
sqrt_list = [ n**2 for n in range(1,21)]
print(sqrt_list)
```

```python
# filtering even numbers in range
num_range = range(1,21)
even_nums = []
srt_list = []
for n in num_range:
  if n%2 == 0:
    even_nums.append(n)
print(even_nums)

# after using list comprehension
print('Using list comprhension')
even_nums = [n for n in num_range if n%2==0]
print(even_nums)

# printing sqrt of even nums
print('sqrt of even nums')
sqrt_list = [n**2 for n in even_nums]
print(sqrt_list)
```
