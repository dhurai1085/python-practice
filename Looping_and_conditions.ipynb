```python
num = input('Enter the numbers :')
my_list = list(map(int, (num).split()))
print(my_list)
for x in my_list:
  if x % 2 == 0 :
    print(f'{x} : is even number')
  else:
    print(f'{x} : is odd number')
```

```python
# Problem 2 — Count positives and negatives
num = input('Enter a number :')
p = 0
n = 0
my_list = list(map(int, num.split()))
for x in my_list:
  if x > 0:
    p += 1
  elif x < 0:
    n += 1
print(f'positive numbers in the list : {p}')
print(f'negative numbers in the list : {n}')
```

```python
# Problem 1: List + Input + Loop + Condition

x = int(input('How many numbers you want to enter:'))
my_list = []
even_list = []
odd_list =[]
for i in range(x):
  num = int(input(f'Enter {i+1} number : '))
  my_list.append(num)
print(my_list)
for y in my_list:
  if y%2 == 0:
    even_list.append(y)
  elif y !=0:
    odd_list.append(y)
print(f"Even List : {even_list}")
print(f"Odd List : {odd_list}")
print('Program End...')
```

```python
# Problem 3: Tuples + Condition
passed = 0
failed = 0
marks = (45, 67, 82, 30, 76, 50)
for x in marks:
  if x >= 50:
    passed += 1
  else:
    failed += 1
print(f'passed students :{passed}')
print(f' Failed students : {failed}')
```

```python
"""🔹 Problem 4: Sets (Duplicates + Logic)
👉 Task:
Take input from user:
Ask how many numbers
Store them in a list
Convert it into a set
Print:
Original list
Unique values (set)
Count of duplicates removed """
my_set = set()
my_list = []

#getting the user input
n = int(input('How many numbers u want to enter:'))
for i in range(n):
  num = int(input('Enter number :'))
  my_list.append(num)

my_set = set(my_list)
# counting the duplicates
dupes = 0
for x in my_set:
  dupes += my_list.count(x)-1
# print the list, set and count of duplictes
print(f'Original list: {my_list}')
print(f'Unique set: {my_set}')
print(f'Duplicates: {dupes}')

```

```python
# Problem 5: Dictionary + Input + Condition
""" 👉 Task:
Ask user: how many students
For each student:
Take name
Take marks
Store in a dictionary """

student_count = int(input('Enter the student count:'))
student_record = {}
for x in range (student_count) :
  key = input('Enter the key:')
  value = int(input('Enter a value:'))
  student_record[key] = value
print(student_record)
print(f'Number of students : {student_count}')

for name in student_record :
  x = name.value
  print(x)
```

```python
text = input('enter a text').lower()

vowels = consonant = digit = specialcharacter = 0
for ch in text:
  if ch in 'aeiou':
    vowels += 1
  elif ch.isalpha():
    consonant += 1
  elif ch.isdigit():
    digit += 1
  else:
    specialcharacter += 1
print(vowels, end = ' ')
print(consonant, end = ' ')
print(digit, end = ' ')
print(specialcharacter, end = ' ')
```
