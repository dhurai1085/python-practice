```python
#  Write a program to check if a given number is a palindrome
# (reads the same forwards and backwards).

num1 = 121
num2 = 125
palindrome = [num1,num2]
for n in palindrome:
  x = str(n)
  if str(n) == x[::-1]:
    print(f'{n} is palindrome')
  else:
    print(f'{n} is not palindrome')
```
